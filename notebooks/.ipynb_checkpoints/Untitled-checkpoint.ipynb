```python
import pandas as pd
from sys import path
path.append('/Users/ridleyleisy/Documents/nfl/database/data/')
from db_getter import dbGetter
```

```python
db = dbGetter()
```

```python
df = pd.DataFrame(db.query('''SELECT *
FROM games as g 
LEFT JOIN teams as t
ON g.gid = t.gid
WHERE g.v = 'ATL' OR g.h = 'ATL' '''))
```

```python
tcols = [val[0] for val in db.query('DESCRIBE teams')]
gcols = [val[0] for val in db.query('DESCRIBE games')]
```

```python
gcols.extend(tcols)
```

```python
df.columns = gcols
```

```python
df.shape
```

```python
def coaching(col):
    if col <= 2011:
        coach = 'Mike_Mularkey'
        
    elif col <= 2014:
        coach = 'Dirk_Koetter'
        
    elif col <= 2016:
        coach = 'Kyle_Shanahan'
        
    elif col <= 2018:
        coach = 'Steve_Sarkisian'
        
    return coach
```

```python
df['coach'] = df['seas'].apply(coaching)
```

```python
df = df.drop_duplicates(subset='gid')
```
